___EXERCICIO 1___

```python
#importando librerias
import pandas as pd
import numpy as np


# importando csv
df = pd.read_csv('DelayedFlights.csv', sep=",", dayfirst=True)

# eliminando primera columna
df = df.drop('Unnamed: 0', 1)

#visualizando formato de las columnas y tamaño de la base
df.info()
df.shape
```

```python
# contenido de la base
df.head()
```

```python
#eliminando columnas no relevantes
df = df.drop(['TaxiIn','TaxiOut','Cancelled','CancellationCode','Diverted','WeatherDelay','NASDelay','SecurityDelay','LateAircraftDelay'], axis=1)

df
```

___EJERCICIO 2___

```python
#Resumen estadistico
df.describe()
```

```python
#datos que faltan por columna
df.isnull().sum()
```

```python
#columna nuevas de velocidad media
df['velocidad_media']=round(df['Distance']/df['AirTime'],2)

df.groupby('UniqueCarrier')[['UniqueCarrier','velocidad_media']].head(5)

```

```python
#columna nueva para ver si ha llegado tarde o no
df['retraso']=df['ArrDelay']>0

df.groupby('UniqueCarrier')[['UniqueCarrier','retraso']].head(5)
```

```python
#Aerolineas con más retrasos
df.groupby('UniqueCarrier').sum()['retraso'].sort_values(ascending=False).head(10)
```

```python
#Vuelos más largos
df.groupby('FlightNum').sum()['retraso'].sort_values(ascending=False).head(10)
```

```python
#Vuelos con más retrasos
df.groupby('FlightNum').mean()['AirTime'].sort_values(ascending=False).head(10)
```

```python
#exportando el nuevo data set

df.to_csv('DelayedFlightsv3.csv')
```
